### IMPORT

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import json 
```

### LOAD DATA

```python
query = "rumah sakit Yogyakarta"
url = f"https://nominatim.openstreetmap.org/search?q={query}&format=json&limit=100"
headers = {
    'User-Agent': 'Datathon-Bintoro-Project/1.0 (email@example.com)'
}
```

```python
response = requests.get(url, headers=headers)
data = response.json()
print(json.dumps(data, indent=2))
```

```python
results = []
for place in data:
    # Memastikan 'display_name' ada, dimulai dengan "Rumah Sakit", DAN 'type' adalah "hospital"
    if place.get('display_name') and \
       place.get('display_name').startswith("Rumah Sakit") and \
       place.get('type') == 'hospital': # Kondisi baru ditambahkan di sini
        results.append({
            'name': place.get('display_name'),
            'latitude': place.get('lat'),
            'longitude': place.get('lon'),
            'type': place.get('type')
        })
```

```python
df = pd.DataFrame(results)
df.to_csv('rumah_sakit_yogyakarta.csv', index=False)
print("✅ Data berhasil disimpan ke rumah_sakit_yogyakarta.csv")
```
